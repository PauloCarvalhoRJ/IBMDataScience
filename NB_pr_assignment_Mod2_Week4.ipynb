### The following code simply demonstrates how to implement recursion in Python (the classic factorial function):

```python
def factorial(n):
   if n == 1:
       print("Returning 1! = 1 (trivial case).")
       return 1 #The trivial case is the end-of-recursion condition.
   else:
       retval = n * factorial( n - 1 ) # recursion rationale: 3! == 3 * 2! == 3 * 2 * 1! == 3 * 2 * 1
       print("Returning",n,"*",n-1,"! = ",retval)
       return retval  
    
print( factorial(3) )
```
